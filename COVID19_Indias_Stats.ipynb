```python
# importing required libraries

import requests
from bs4 import BeautifulSoup
```

```python
# getting data from indian govt. portal
page = requests.get("https://www.mygov.in/corona-data/covid19-statewise-status")

page.status_code # checking for successful connection
```

```python
# creating BeautifulSoup v4's object. The default parser is html.
bs = BeautifulSoup(page.text)
```

```python
# getting required tags using Beautiful Soup's 'find_all()' method
t = bs.find_all('div', attrs={'class':'field-item even'})[:6]

# extracting the required data from tags returned by 'find_all() method'
last_up = t[0].text
total_cases = int(t[2].text)+int(t[3].text)+int(t[4].text)+int(t[5].text)
total_recovered = t[3].text
total_deaths = t[5].text

# 'summary' string to make the output presentable 
summary = f'''
        Data Last Updated on: {last_up}
        
        Total Confirmed Cases: {total_cases}
        Recovered: {total_recovered}
        Deaths: {total_deaths}
        '''
```

```python
country_data = [] # this list holds the data of all the states/UTs

# getting required tags using Beautiful Soup's 'find_all()' method
for row in bs.find_all('div', attrs={'class':'content'})[3:35]:
    
    state_data = [] # will be used to store the data of each state/UT
    
    # extracting the required data from tags
    raw_stats = row.text.splitlines()[1].split('\xa0')[1:]
    state = raw_stats[0].split('To')[0]
    confirmed = int(raw_stats[1].split('C')[0])
    recovered = int(raw_stats[2].split('D')[0])
    deaths = int(raw_stats[3].split()[0])
    
    # adding the data to 'state_data' list
    state_data.append(state)
    state_data.append(confirmed)
    state_data.append(recovered)
    state_data.append(deaths)
    
    # finally adding the state's data to the list containing data from all the states/UTs
    country_data.append(state_data)
```

```python
import pandas as pd # importing pandas library to use its DataFrames

# creating pandas DataFrame object to facilitate analysis of data
df = pd.DataFrame(country_data, columns=['State','Confirmed','Recovered','Deaths'])
df['S.No.'] = df.index + 1
df.set_index('S.No.', inplace=True)
```

```python
# printing the data obtained from the website
print(summary)
df
```
